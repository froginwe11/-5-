# reflow

英\[ˈri:fləʊ\] 美\[ˈrifloʊ\]

**[释义](https://fanyi.baidu.com/#en/zh/reflow)**

n. 软熔焊接; 文档重整，页面重排（调整文本在计算机屏幕上的显示密度）

**例句**：

After reflow, carefully remove substrate with components and allow to cool.
